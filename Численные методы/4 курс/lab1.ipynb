Используя явную и неявную конечно-разностные схемы, а также схему Кранка - Николсона, решить начально-краевую задачу для дифференциального уравнения параболического типа. Осуществить реализацию трех вариантов аппроксимации граничных условий, содержащих производные: двухточечная аппроксимация с первым порядком, трехточечная аппроксимация со вторым порядком, двухточечная аппроксимация со вторым порядком. В различные моменты времени вычислить погрешность численного решения путем сравнения результатов с приведенным в задании аналитическим решением $U(x, t)$. Исследовать зависимость погрешности от сеточных параметров  $\tau, h$.

Общий вид задачи:

$\dfrac{du}{dt} = a\dfrac{d^2u}{dx^2} + b\dfrac{du}{dx} + cu + f(x, t)$ <br>

$\alpha_{0}u'_{x}(0, t) + \beta_{0}u(0, t) = \phi_{0}(t)$ <br>
$\alpha_{1}u'_{x}(l, t) + \beta_{1}u(l, t) = \phi_{1}(t)$ <br>
$u(x, 0) = \psi_{0}(x)$ <br>

**Для решения задачи используются 3 вида конечно-разностных схем:**
**1. явная схема**

$\dfrac{u^{k+1}_i - u^k_i}{\tau} = a\dfrac{u^{k}_{i-1} - 2u^{k}_{i} + u^{k}_{i+1}}{h^2} + b\dfrac{u^{k}_{i+1} - u^{k}_{i-1}}{2h} + cu^{k}_{i} + f(x_i, t^{k+1})$

**2. неявная схема**

$\dfrac{u^{k+1}_i - u^k_i}{\tau} = a\dfrac{u^{k+1}_{i-1} - 2u^{k+1}_{i} + u^{k+1}_{i+1}}{h^2} + b\dfrac{u^{k+1}_{i+1} - u^{k+1}_{i-1}}{2h} + cu^{k+1}_{i} + f(x_i, t^k)$

**3. схема Кранка-Николсона** ($\theta = 0.5$)

$\dfrac{u^{k+1}_i - u^k_i}{\tau} = \theta\bigg(a\dfrac{u^{k+1}_{i-1} - 2u^{k+1}_{i} + u^{k+1}_{i+1}}{h^2} + b\dfrac{u^{k+1}_{i+1} - u^{k+1}_{i-1}}{2h} + cu^{k+1}_{i} + f(x_i, t^k)\bigg) + (1 - \theta)\bigg(a\dfrac{u^{k}_{i-1} - 2u^{k}_{i} + u^{k}_{i+1}}{h^2} + b\dfrac{u^{k}_{i+1} - u^{k}_{i-1}}{2h} + cu^{k}_{i} + f(x_i, t^{k+1})\bigg)$

**Для решения задачи используются 3 вида аппроксимации граничных условий:**
**1. двухточечная аппроксимация с первым порядком**

$\dfrac{du}{dx}\bigg|^{k+1}_{j = 0} = \dfrac{u^{k+1}_1 - u^{k+1}_0}{h}$

$\dfrac{du}{dx}\bigg|^{k+1}_{j = N} = \dfrac{u^{k+1}_N - u^{k+1}_{N-1}}{h}$

Подставим в граничные условия и получим следующие формулы

$(h\beta_0 - \alpha_0)u^{k+1}_0 + \alpha_0u^{k+1}_1 = h\phi_0(t^{k+1})$ <br>

$-\alpha_1u^{k+1}_{N-1} + (\alpha_1 + h\beta_1)u^{k+1}_{N} = h\phi_1(t^{k+1})$

**2. трехтоточная аппроксимация со вторым порядком**

$\dfrac{du}{dx}\bigg|^{k+1}_{j = 0} = \dfrac{-3u^{k+1}_0 + 4u^{k+1}_1 - u^{k+1}_2}{2h}$

$\dfrac{du}{dx}\bigg|^{k+1}_{j = N} = \dfrac{u^{k+1}_{N - 2} - 4u^{k+1}_{N - 1} + 3u^{k+1}_N}{2h}$

Подставим в граничные условия и получим следующие формулы

$(-3\alpha_0 + 2h\beta_0)u^{k+1}_0 + 4\alpha_0u^{k+1}_1 - \alpha_0 u^{k+1}_2 = 2h\phi_0(t^{k+1})$

$\alpha_1u^{k+1}_{N-2} - 4\alpha_1u^{k+1}_{N-1} + (3\alpha_1 + 2h\beta_1)u^{k+1}_N = 2h\phi_1(t^{k+1})$

**3. двухточечная аппроксимация со вторым порядком**

Разложим в граничных узлах на точном решении значение $u^{k+1}_1$ и $u^{k+1}_{N-1}$ в окрестности точки $x=0$ в ряд Тейлора по переменной $x$ до третьей производной включительно

$u^{k+1}_1 = u(0 + h, t^{k + 1}) = u^{k+1}_0 + \dfrac{du}{dx}\bigg|^{k + 1}_{0}\cdot h + \dfrac{d^2u}{dx^2}\bigg|^{k + 1}_{0}\cdot\dfrac{h^2}{2}$

$u^{k+1}_{N-1} = u(l - h, t^{k + 1}) = u^{k+1}_N - \dfrac{du}{dx}\bigg|^{k + 1}_{N}\cdot h + \dfrac{d^2u}{dx^2}\bigg|^{k + 1}_{N}\cdot\dfrac{h^2}{2}$

Выразим значение второй производной по $x$ из дифференциального уравнения

$\dfrac{d^2u}{dx^2}\bigg|^{k + 1}_{j} = (\dfrac{1}{a}\cdot\dfrac{du}{dt} - \dfrac{b}{a}\cdot\dfrac{du}{dx} - \dfrac{c}{a}u - \dfrac{f(x, t)}{a})$

Подставим в разложение и выразим $\dfrac{du}{dx}\bigg|^{k + 1}_{0}$ и $\dfrac{du}{dx}\bigg|^{k + 1}_{N}$

$\dfrac{du}{dx}\bigg|^{k + 1}_{0} = \dfrac{u_1 - u_0}{h} - \dfrac{d^2u}{dx^2}\bigg|^{k + 1}_{0}\cdot\dfrac{h}{2} = \dfrac{u_1 - u_0}{h} - (\dfrac{1}{a}\cdot\dfrac{du}{dt}\bigg|^{k + 1}_{0} - \dfrac{b}{a}\cdot\dfrac{du}{dx}\bigg|^{k + 1}_{0} - \dfrac{c}{a}u_0 - \dfrac{f(x, t)}{a})\dfrac{h}{2}$

$\dfrac{du}{dx}\bigg|^{k + 1}_{N} = \dfrac{u_N - u_{N-1}}{h} + \dfrac{d^2u}{dx^2}\bigg|^{k + 1}_{N}\cdot\dfrac{h}{2} = \dfrac{u_N - u_{N-1}}{h} + (\dfrac{1}{a}\cdot\dfrac{du}{dt}\bigg|^{k + 1}_{N} - \dfrac{b}{a}\cdot\dfrac{du}{dx}\bigg|^{k + 1}_{N} - \dfrac{c}{a}u_N - \dfrac{f(x, t)}{a})\dfrac{h}{2}$

$(1 - \dfrac{bh}{2a})\dfrac{du}{dx}\bigg|^{k + 1}_{0} = \dfrac{u_1 - u_0}{h} - (\dfrac{1}{a}\cdot\dfrac{du}{dt}\bigg|^{k + 1}_{0} - \dfrac{c}{a}u_0 - \dfrac{f(x, t)}{a})\dfrac{h}{2}$

$(1 + \dfrac{bh}{2a})\dfrac{du}{dx}\bigg|^{k + 1}_{N} = \dfrac{u_N - u_{N-1}}{h} + (\dfrac{1}{a}\cdot\dfrac{du}{dt}\bigg|^{k + 1}_{N} - \dfrac{c}{a}u_N - \dfrac{f(x, t)}{a})\dfrac{h}{2}$

Подставляя аппроксимации производных из основного уравнения и подставив выражение в граничные условия получим следующие формулы:

$\alpha_0(-2a\tau - h^2 + c\tau h^2 + \dfrac{\beta_0}{\alpha_0}(2a\tau h - bh^2\tau))u^{k+1}_0 + \alpha_0(2a\tau)u^{k+1}_1 = \phi_0(2a\tau h - bh^2\tau) + \alpha_0(-u^{k-1}_0 h^2 - f(0, t)\tau h^2)$

$\alpha_1(2a\tau + h^2 - c\tau h^2 + \dfrac{\beta_1}{\alpha_1}(2a\tau h + bh^2\tau))u^{k+1}_N + \alpha_0(-2a\tau)u^{k+1}_{N-1} = \phi_1(2a\tau h + bh^2\tau) + \alpha_1(u^{k-1}_N h^2 + f(l, t)\tau h^2)$

```python
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
params = {
    'a': 1.0,
    'b': 0.0,
    'c': 0.0,
    'f': lambda x, t: 0.0,
    'alpha_0': 1.0,
    'beta_0': 0.0,
    'f_phi_0': lambda t: np.exp(-t),
    'alpha_1': 1.0,
    'beta_1': 0.0,
    'f_phi_1': lambda t: -np.exp(-t),
    'f_psi': lambda x: np.sin(x),
    'interval': (0.0, np.pi),
    'analytical_solution': lambda x, t: np.exp(-t) * np.sin(x)
}
```

```python
class Task:
    def __init__(self, **params):
        self.a = params['a']
        self.b = params['b']
        self.c = params['c']
        self.f = params['f']
        
        self.f_psi = params['f_psi']
        self.interval = params['interval']
        
        self.alpha_0 = params['alpha_0']
        self.beta_0 = params['beta_0']
        self.f_phi_0 = params['f_phi_0']
        
        self.alpha_1 = params['alpha_1']
        self.beta_1 = params['beta_1']
        self.f_phi_1 = params['f_phi_1']
        
        self.analytical_solution = params['analytical_solution']
        
class Solver:
    
    @classmethod
    def solve(cls, task:Task, n, max_time, scheme_type, bound_appr_type):
        h, tau, time_cnt = cls.__prepare(task.interval[0], task.interval[1], task.a, n, max_time)
        if scheme_type == 'explicit':
            u = cls.__explicit_solve(task, h, tau, n, time_cnt, bound_appr_type)
        elif scheme_type == 'implicit':
            u = cls.__implicit_solve(task, h, tau, n, time_cnt, bound_appr_type)
        elif scheme_type == 'crank_nicolson':
            u = cls.__crank_nicolson_solve(task, h, tau, n, time_cnt, bound_appr_type)
        x = cls.__get_split_by_space(h, n)
        t = cls.__get_split_by_time(tau, time_cnt)
        return u, x, t
    
    @classmethod
    def __explicit_solve(cls, task:Task, h, tau, n, time_cnt, bound_appr_type):
        u = np.zeros((time_cnt, n))
        z_0 = z_1 = b_0 = c_0 = a_n = b_n = 0.0
        if bound_appr_type == '2p2a':
            z_0 = 2 * task.a * tau * h - task.b * h ** 2 * tau
            z_1 = 2 * task.a * tau * h + task.b * h ** 2 * tau
            b_0 = task.alpha_0 * (-2 * task.a * tau - h ** 2 + tau * h ** 2 * task.c) + task.beta_0 * z_0
            c_0 = 2 * task.alpha_0 * task.a * tau
            a_n = -2 * task.alpha_1 * task.a * tau
            b_n = task.alpha_1 * (2 * task.a * tau + h ** 2 - tau * h ** 2 * task.c) + task.beta_1 * z_1
        
        for j in range(n):
            u[0][j] = task.f_psi(h * j)
            
        for k in range(1, time_cnt):
            for j in range(1, n - 1):
                u[k][j] = (2 * task.a * tau * (u[k - 1][j - 1] - 2 * u[k - 1][j] + u[k - 1][j + 1]) + \
                          task.b * h * tau * (u[k - 1][j + 1] - u[k - 1][j - 1]) + \
                          2 * task.c * h ** 2 * tau * u[k - 1][j] + 2 * h ** 2 * tau * task.f(h * j, tau * (k - 1)) + \
                          2 * h ** 2 * u[k - 1][j]) / (2 * h ** 2)
            if bound_appr_type == '2p1a':
                u[k][0] = (h * task.f_phi_0(tau * k) - task.alpha_0 * u[k][1]) / (h * task.beta_0 - task.alpha_0)
                u[k][-1] = (h * task.f_phi_1(tau * k) + task.alpha_1 * u[k][-2]) / (h * task.beta_1 + task.alpha_1)
            elif bound_appr_type == '3p2a':
                u[k][0] = (2 * h * task.f_phi_0(tau * k) + task.alpha_0 * u[k][2] - 4 * task.alpha_0 * u[k][1]) / \
                          (-3 * task.alpha_0 + 2 * h * task.beta_0)
                u[k][-1] = (2 * h * task.f_phi_1(tau * k) - task.alpha_1 * u[k][-3] + 4 * task.alpha_1 * u[k][-2]) / \
                           (3 * task.alpha_1 + 2 * h * task.beta_1)
            elif bound_appr_type == '2p2a':
                d_0 = task.f_phi_0(tau * k) * z_0 + task.alpha_0 * (-u[k - 1][0] * h ** 2 - tau * h ** 2 * task.f(0.0, tau * k))
                u[k][0] = (d_0 - c_0 * u[k][1]) / b_0
                d_n = task.f_phi_1(tau * k) * z_1 + task.alpha_1 * (u[k - 1][-1] * h ** 2 + tau * h ** 2 * task.f(task.interval[1], tau * k))
                u[k][-1] = (d_n - a_n * u[k][-2]) / b_n
        return u
    
    @classmethod
    def __implicit_solve(cls, task:Task, h, tau, n, time_cnt, bound_appr_type):
        u = np.zeros((time_cnt, n))
        d = np.zeros((time_cnt, n))
        
        a = np.array([0.0] + [-2 * tau * task.a + h * tau * task.b for _ in range(1, n - 1)] + [0.0])
        b = np.array([0.0] + [2 * h ** 2 + 4 * tau * task.a - 2 * h ** 2 * tau * task.c for _ in range(1, n - 1)] + [0.0])
        c = np.array([0.0] + [-2 * tau * task.a - h * tau * task.b for _ in range(1, n - 1)] + [0.0])
        
        if bound_appr_type == '2p1a':
            b[0] = h * task.beta_0 - task.alpha_0
            c[0] = task.alpha_0
            a[-1] = -task.alpha_1
            b[-1] = task.alpha_1 + h * task.beta_1
        elif bound_appr_type == '3p2a':
            b[0] = -3 * task.alpha_0 + 2 * h * task.beta_0 + a[1] * task.alpha_0 / (-2 * tau * task.a - h * tau * task.b)
            c[0] = 4 * task.alpha_0 + b[1] * task.alpha_0 / (-2 * tau * task.a - h * tau * task.b)
            a[-1] = -4 * task.alpha_1 - b[1] * task.alpha_1 / (-2 * tau * task.a + h * tau * task.b)
            b[-1] = 3 * task.alpha_1 + 2 * h * task.beta_1 - c[1] * task.alpha_1 / (-2 * tau * task.a + h * tau * task.b)
        elif bound_appr_type == '2p2a':
            z_0 = 2 * task.a * tau * h - task.b * h ** 2 * tau
            z_1 = 2 * task.a * tau * h + task.b * h ** 2 * tau
            b[0] = task.alpha_0 * (-2 * task.a * tau - h ** 2 + tau * h ** 2 * task.c) + task.beta_0 * z_0
            c[0] = 2 * task.alpha_0 * task.a * tau
            a[-1] = -2 * task.alpha_1 * task.a * tau
            b[-1] = task.alpha_1 * (2 * task.a * tau + h ** 2 - tau * h ** 2 * task.c) + task.beta_1 * z_1
            
        for j in range(n):
            u[0][j] = task.f_psi(h * j)
            
        for k in range(1, time_cnt):
            for j in range(1, n - 1):
                d[k][j] = 2 * h ** 2 * u[k - 1][j] + 2 * h ** 2 * tau * task.f(h * j, tau * k)
            if bound_appr_type == '2p1a':
                d[k][0] = h * task.f_phi_0(tau * k)
                d[k][-1] = h * task.f_phi_1(tau * k)
            elif bound_appr_type == '3p2a':
                d[k][0] = 2 * h * task.f_phi_0(tau * k) + d[k][1] * task.alpha_0 / (-2 * tau * task.a - h * tau * task.b)
                d[k][-1] = 2 * h * task.f_phi_1(tau * k) - d[k][-2] * task.alpha_1 / (-2 * tau * task.a + h * tau * task.b)
            elif bound_appr_type == '2p2a':
                d[k][0] = task.f_phi_0(tau * k) * z_0 + task.alpha_0 * (-u[k - 1][0] * h ** 2 - tau * h ** 2 * task.f(0.0, tau * k))
                d[k][-1] = task.f_phi_1(tau * k) * z_1 + task.alpha_1 * (u[k - 1][-1] * h ** 2 + tau * h ** 2 * task.f(task.interval[1], tau * k))
            u_k = cls.__tma(a, b, c, d, n, k)
            for j in range(n):
                u[k][j] = u_k[j]
        return u
    
    @classmethod
    def __crank_nicolson_solve(cls, task:Task, h, tau, n, time_cnt, bound_appr_type):
        u = np.zeros((time_cnt, n))
        d = np.zeros((time_cnt, n))
        
        a = np.array([0.0] + [-2 * tau * task.a * cls.__theta + h * tau * task.b * cls.__theta for _ in range(1, n - 1)] + [0.0])
        b = np.array([0.0] + [2 * h ** 2 + 4 * tau * task.a * cls.__theta - 2 * h ** 2 * tau * task.c * cls.__theta for _ in range(1, n - 1)] + [0.0])
        c = np.array([0.0] + [-2 * tau * task.a * cls.__theta - h * tau * task.b * cls.__theta for _ in range(1, n - 1)] + [0.0])
        
        if bound_appr_type == '2p1a':
            b[0] = h * task.beta_0 - task.alpha_0
            c[0] = task.alpha_0
            a[-1] = -task.alpha_1
            b[-1] = task.alpha_1 + h * task.beta_1
        elif bound_appr_type == '3p2a':
            b[0] = -3 * task.alpha_0 + 2 * h * task.beta_0 + a[1] * task.alpha_0 / (-2 * tau * task.a * cls.__theta - h * tau * task.b * cls.__theta)
            c[0] = 4 * task.alpha_0 + b[1] * task.alpha_0 / (-2 * tau * task.a * cls.__theta - h * tau * task.b * cls.__theta)
            a[-1] = -4 * task.alpha_1 - b[1] * task.alpha_1 / (-2 * tau * task.a * cls.__theta + h * tau * task.b * cls.__theta)
            b[-1] = 3 * task.alpha_1 + 2 * h * task.beta_1 - c[1] * task.alpha_1 / (-2 * tau * task.a * cls.__theta + h * tau * task.b * cls.__theta)
        elif bound_appr_type == '2p2a':
            z_0 = 2 * task.a * tau * h * cls.__theta - task.b * h ** 2 * tau * cls.__theta
            z_1 = 2 * task.a * tau * h * cls.__theta + task.b * h ** 2 * tau * cls.__theta
            b[0] = (task.alpha_0 * (-2 * task.a * tau * cls.__theta - h ** 2 + tau * h ** 2 * task.c * cls.__theta) + task.beta_0 * z_0) * task.alpha_0
            c[0] = (2 * task.alpha_0 * task.a * tau * cls.__theta) * task.alpha_0
            a[-1] = (-2 * task.alpha_1 * task.a * tau * cls.__theta) * task.alpha_1
            b[-1] = (task.alpha_1 * (2 * task.a * tau * cls.__theta + h ** 2 - tau * h ** 2 * task.c * cls.__theta) + task.beta_1 * z_1) * task.alpha_1
            
        for j in range(n):
            u[0][j] = task.f_psi(h * j)
            
        for k in range(1, time_cnt):
            for j in range(1, n - 1):
                p = task.a * (u[k - 1][j - 1] - 2 * u[k - 1][j] + u[k - 1][j + 1]) / (h ** 2) + \
                    task.b * (u[k - 1][j + 1] - u[k - 1][j - 1]) / (2 * h) + \
                    task.c * u[k - 1][j] + task.f(h * j, tau * (k - 1))
                d[k][j] = 2 * h ** 2 * u[k - 1][j] + \
                          2 * h ** 2 * tau * task.f(h * j, tau * k) * cls.__theta + \
                          (1 - cls.__theta) * p * 2 * h ** 2 * tau
            if bound_appr_type == '2p1a':
                d[k][0] = h * task.f_phi_0(tau * k)
                d[k][-1] = h * task.f_phi_1(tau * k)
            elif bound_appr_type == '3p2a':
                d[k][0] = 2 * h * task.f_phi_0(tau * k) + d[k][1] * task.alpha_0 / (-2 * tau * task.a * cls.__theta - h * tau * task.b * cls.__theta)
                d[k][-1] = 2 * h * task.f_phi_1(tau * k) - d[k][-2] * task.alpha_1 / (-2 * tau * task.a * cls.__theta + h * tau * task.b * cls.__theta)
            elif bound_appr_type == '2p2a':
                u_m1 = 2 * h * task.beta_0 * u[k - 1][0] + u[k - 1][1] * task.alpha_0 - 2 * h * task.f_phi_0(tau * (k - 1))
                u_p1 = u[k - 1][-2] * task.alpha_1 - 2 * h * task.beta_1 * u[k - 1][-1] + 2 * h * task.f_phi_1(tau * (k - 1))
                p_0 = task.alpha_0 * task.a * (u_m1 - 2 * u[k - 1][0] + u[k - 1][1]) / (h ** 2) + \
                      task.alpha_0 * task.b * (u[k - 1][1] -  u_m1) / (2 * h) + \
                      task.alpha_0 * task.c * u[k - 1][0] +  task.alpha_0 * task.f(0.0, tau * (k - 1))
                p_n = task.alpha_1 * task.a * (u[k - 1][-2] - 2 * u[k - 1][-1] + u_p1) / (h ** 2) + \
                      task.alpha_1 * task.b * (u_p1 - u[k - 1][-2]) / (2 * h) + \
                      task.alpha_1 * task.c * u[k - 1][-1] + task.alpha_1 * task.f(task.interval[1], tau * (k - 1))
                d[k][0] = (task.f_phi_0(tau * k) * z_0 + \
                           task.alpha_0 * (-u[k - 1][0] * h ** 2 - tau * h ** 2 * task.f(0.0, tau * k) * cls.__theta)) * task.alpha_0 - \
                           (1 - cls.__theta) * p_0 * tau * h ** 2 * z_0
                d[k][-1] = (task.f_phi_1(tau * k) * z_1 + \
                            task.alpha_1 * (u[k - 1][-1] * h ** 2 + tau * h ** 2 * task.f(task.interval[1], tau * k) * cls.__theta)) * task.alpha_1 + \
                            (1 - cls.__theta) * p_n * tau * h ** 2 * z_1
            u_k = cls.__tma(a, b, c, d, n, k)
            for j in range(n):
                u[k][j] = u_k[j]
        return u
    
    @classmethod
    def analytical_solution(cls, task:Task, n, max_time):
        h, tau, time_cnt = cls.__prepare(task.interval[0], task.interval[1], task.a, n, max_time)
        res = np.zeros((time_cnt, n))
        for k in range(time_cnt):
            for j in range(n):
                res[k][j] = task.analytical_solution(h * j, tau * k)
        return res
    
    @classmethod
    def __tma(cls, a, b, c, d, n, k):
        A = np.zeros(n)
        B = np.zeros(n)
        x = np.zeros(n)
        A[0] = -c[0] / b[0]
        B[0] = d[k][0] / b[0]
        
        for j in range(1, n):
            A[j] = -c[j] / (b[j] + a[j] * A[j - 1])
            B[j] = (d[k][j] - a[j] * B[j - 1]) / (b[j] + a[j] * A[j - 1])
        x[-1] = B[-1]
        for j in range(n - 2, -1, -1):
            x[j] = A[j] * x[j + 1] + B[j]
        return x
    
    @classmethod
    def __get_split_by_space(cls, h, n):
        return np.array([h * j for j in range(n)])
    
    @classmethod
    def __get_split_by_time(cls, tau, time_cnt):
        return np.array([tau * k for k in range(time_cnt)])
    
    @classmethod
    def __prepare(cls, begin, end, a, n, max_time):
        h = (end - begin) / (n - 1)
        tau = cls.__sigma * h ** 2 / a
        time_cnt = int(max_time / tau)
        return h, tau, time_cnt
    
    __sigma = 0.4
    __theta = 0.5
```
