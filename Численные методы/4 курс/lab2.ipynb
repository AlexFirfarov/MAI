Используя явную схему крест и неявную схему, решить начально-краевую задачу для дифференциального уравнения гиперболического типа. Аппроксимацию второго начального условия произвести с первым и со вторым порядком. Осуществить реализацию трех вариантов аппроксимации граничных условий, содержащих производные: двухточечная аппроксимация с первым порядком, трехточечная аппроксимация со вторым порядком, двухточечная аппроксимация со вторым порядком. В различные моменты времени вычислить погрешность численного решения путем сравнения результатов с приведенным в задании аналитическим решением $U(x, t)$. Исследовать зависимость погрешности от сеточных параметров $\tau, h$.

Общий вид задачи:

$\dfrac{d^2u}{dt^2} + d\dfrac{du}{dt} = a^2\dfrac{d^2u}{dx^2} + b\dfrac{du}{dx} + cu + f(x, t)$ <br>

$\alpha_{0}u'_{x}(0, t) + \beta_{0}u(0, t) = \phi_{0}(t)$ <br>
$\alpha_{1}u'_{x}(l, t) + \beta_{1}u(l, t) = \phi_{1}(t)$ <br>
$u(x, 0) = \psi_{0}(x)$ <br>
$u'_{t}(x, 0) = \psi_{1}(x)$ <br>

Для старта вычислительного процесса необходимо знать значения сеточных функций на двух первых временных слоях $u^0, u^1$ <br>

$u^0_{j} = \psi_{0}(x_{j})$ <br>

Найдем $u^1$ через аппроксимацию второго начального условия

**Аппроксимация с 1 порядком**

Воспользуемся простейшей аппроксимацией второго начального условия <br>

$\dfrac{u^1_{j} - u^0_{j}}{\tau} = \psi_{1}(x_{j})$ <br>

Отсюда для $u^1$ получаем следующее выражение <br>

$u^1_{j} = u^0_{j} + \tau \psi_{1}(x_{j})$

Недостатком такого подхода является первый порядок аппроксимации второго начального условия.

**Аппроксимация со 2 порядком**

Разложим в ряд Тейлора $u^1_{j}$ на точном решении по времени в окрестности $t = 0$ <br>

$u^1_{j} = u(x_{j}, 0 + \tau) = u(x_{j}, 0) + \tau u'_{t}(x_{j}, 0) + \dfrac{\tau^2}{2}u''_{tt}(x_{j}, 0) = 
\psi_{0}(x_{j}) + \tau \psi_{1}(x_{j}) + \dfrac{\tau^2}{2}(a^2\psi''_{0}(x_{j}) + b\psi'_{0}(x_{j}) + c\psi_{0}(x_{j}) + f(x_{j}, 0) - d\psi_{1}(x_{j})) = (1 + \dfrac{c\tau^2}{2})\psi_{0}(x_{j}) + \dfrac{b\tau^2}{2}\psi'_{0}(x_{j}) + (\tau - \dfrac{d\tau^2}{2})\psi_{1}(x_{j}) + \dfrac{a^2\tau^2}{2}\psi''_{0}(x_{j}) + \dfrac{\tau^2}{2}f(x_{j}, 0)$ <br>

**В данной лабораторной работе используется 3 вида аппроксимации граничных условий:**
1. двухточечная аппроксимация с первым порядком
2. трехтоточная аппроксимация со вторым порядком
3. двухточечная аппроксимация со вторым порядком

**Двухточечная аппроксимация с первым порядком**

$\dfrac{du}{dx}\bigg|^{k+1}_{j = 0} = \dfrac{u^{k+1}_1 - u^{k+1}_0}{h}$

$\dfrac{du}{dx}\bigg|^{k+1}_{j = N} = \dfrac{u^{k+1}_N - u^{k+1}_{N-1}}{h}$

Подставим в граничные условия и получим следующие формулы

$(h\beta_0 - \alpha_0)u^{k+1}_0 + \alpha_0u^{k+1}_1 = h\phi_0(t^{k+1})$ <br>

$-\alpha_1u^{k+1}_{N-1} + (\alpha_1 + h\beta_1)u^{k+1}_{N} = h\phi_1(t^{k+1})$

**Трехтоточная аппроксимация со вторым порядком**

$\dfrac{du}{dx}\bigg|^{k+1}_{j = 0} = \dfrac{-3u^{k+1}_0 + 4u^{k+1}_1 - u^{k+1}_2}{2h}$

$\dfrac{du}{dx}\bigg|^{k+1}_{j = N} = \dfrac{u^{k+1}_{N - 2} - 4u^{k+1}_{N - 1} + 3u^{k+1}_N}{2h}$

Подставим в граничные условия и получим следующие формулы

$(-3\alpha_0 + 2h\beta_0)u^{k+1}_0 + 4\alpha_0u^{k+1}_1 - \alpha_0 u^{k+1}_2 = 2h\phi_0(t^{k+1})$

$\alpha_1u^{k+1}_{N-2} - 4\alpha_1u^{k+1}_{N-1} + (3\alpha_1 + 2h\beta_1)u^{k+1}_N = 2h\phi_1(t^{k+1})$

**Двухточечная аппроксимация со вторым порядком**

Разложим в граничных узлах на точном решении значение $u^{k+1}_1$ и $u^{k+1}_{N-1}$ в окрестности точки $x=0$ в ряд Тейлора по переменной $x$ до третьей производной включительно

$u^{k+1}_1 = u(0 + h, t^{k + 1}) = u^{k+1}_0 + \dfrac{du}{dx}\bigg|^{k + 1}_{0}\cdot h + \dfrac{d^2u}{dx^2}\bigg|^{k + 1}_{0}\cdot\dfrac{h^2}{2}$

$u^{k+1}_{N-1} = u(l - h, t^{k + 1}) = u^{k+1}_N - \dfrac{du}{dx}\bigg|^{k + 1}_{N}\cdot h + \dfrac{d^2u}{dx^2}\bigg|^{k + 1}_{N}\cdot\dfrac{h^2}{2}$

Выразим значение второй производной по $x$ из дифференциального уравнения

$\dfrac{d^2u}{dx^2}\bigg|^{k + 1}_{j} = (\dfrac{1}{a^2}\cdot\dfrac{d^2u}{dt^2} - \dfrac{b}{a^2}\cdot\dfrac{du}{dx} - \dfrac{c}{a^2}u - \dfrac{f(x, t)}{a^2} + \dfrac{d}{a^2}\cdot\dfrac{du}{dt})$
 
Подставим в разложение и выразим $\dfrac{du}{dx}\bigg|^{k + 1}_{0}$ и $\dfrac{du}{dx}\bigg|^{k + 1}_{N}$

$\dfrac{du}{dx}\bigg|^{k + 1}_{0} = \dfrac{u_1 - u_0}{h} - \dfrac{d^2u}{dx^2}\bigg|^{k + 1}_{0}\cdot\dfrac{h}{2} = \dfrac{u_1 - u_0}{h} - (\dfrac{1}{a^2}\cdot\dfrac{d^2u}{dt^2}\bigg|^{k + 1}_{0} - \dfrac{b}{a^2}\cdot\dfrac{du}{dx}\bigg|^{k + 1}_{0} - \dfrac{c}{a^2}u_0 - \dfrac{f(x, t)}{a^2} + \dfrac{d}{a^2}\cdot\dfrac{du}{dt}\bigg|^{k + 1}_{0})\dfrac{h}{2}$

$\dfrac{du}{dx}\bigg|^{k + 1}_{N} = \dfrac{u_N - u_{N-1}}{h} + \dfrac{d^2u}{dx^2}\bigg|^{k + 1}_{N}\cdot\dfrac{h}{2} = \dfrac{u_1 - u_0}{h} + (\dfrac{1}{a^2}\cdot\dfrac{d^2u}{dt^2}\bigg|^{k + 1}_{N} - \dfrac{b}{a^2}\cdot\dfrac{du}{dx}\bigg|^{k + 1}_{N} - \dfrac{c}{a^2}u_N - \dfrac{f(x, t)}{a^2} + \dfrac{d}{a^2}\cdot\dfrac{du}{dt}\bigg|^{k + 1}_{N})\dfrac{h}{2}$

$(1 - \dfrac{bh}{2a^2})\dfrac{du}{dx}\bigg|^{k + 1}_{0} = \dfrac{u_1 - u_0}{h} - (\dfrac{1}{a^2}\cdot\dfrac{d^2u}{dt^2}\bigg|^{k + 1}_{0} - \dfrac{c}{a^2}u_0 - \dfrac{f(x, t)}{a^2} + \dfrac{d}{a^2}\cdot\dfrac{du}{dt}\bigg|^{k + 1}_{0})\dfrac{h}{2}$

$(1 + \dfrac{bh}{2a^2})\dfrac{du}{dx}\bigg|^{k + 1}_{N} = \dfrac{u_N - u_{N-1}}{h} + (\dfrac{1}{a^2}\cdot\dfrac{d^2u}{dt^2}\bigg|^{k + 1}_{N} - \dfrac{c}{a^2}u_0 - \dfrac{f(x, t)}{a^2} + \dfrac{d}{a^2}\cdot\dfrac{du}{dt}\bigg|^{k + 1}_{N})\dfrac{h}{2}$

Подставляя аппроксимации производных из основного уравнения и подставив выражение в граничные условия получим следующие формулы:

**Явная схема**

$\alpha_0(-2a^2\tau^2-h^2+c\tau^2h^2-0.5d\tau h^2 + \dfrac{\beta_0}{\alpha_0}(2a^2\tau^2h - bh^2\tau^2))u^{k+1}_0 + \alpha_0(2a^2\tau^2)u^{k+1}_1 = \phi_0(2a^2\tau^2h - bh^2\tau^2) + \alpha_0(-2u^{k}_0 h^2 + u^{k-1}_0 h^2 - f(0, t)\tau^2h^2 - 0.5du^{k-1}_0\tau h^2)$

$-\alpha_1(2a^2\tau^2)u^{k+1}_{N-1} + \alpha_1(2a^2\tau^2+h^2-c\tau^2h^2+0.5d\tau h^2 + \dfrac{\beta_1}{\alpha_1}(2a^2\tau^2h + bh^2\tau^2))u^{k+1}_{N} = \phi_1(2a^2\tau^2h + bh^2\tau^2) + \alpha_1(2u^{k}_N h^2 - u^{k-1}_N h^2 + f(l, t)\tau^2h^2 + 0.5du^{k-1}_N\tau h^2)$

**Неявная схема**

$\alpha_0(-2a^2\tau^2-h^2+c\tau^2h^2-1.5d\tau h^2 + \dfrac{\beta_0}{\alpha_0}(2a^2\tau^2h - bh^2\tau^2))u^{k+1}_0 + \alpha_0(2a^2\tau^2)u^{k+1}_1 = \phi_0(2a^2\tau^2h - bh^2\tau^2) + \alpha_0(-2u^{k}_0 h^2 + u^{k-1}_0 h^2 - f(0, t)\tau^2h^2 - 0.5d(u^{k-1}_0 - 4u^{k}_0)\tau h^2)$

$-\alpha_1(2a^2\tau^2)u^{k+1}_{N-1} + \alpha_1(2a^2\tau^2+h^2-c\tau^2h^2+1.5d\tau h^2 + \dfrac{\beta_1}{\alpha_1}(2a^2\tau^2h + bh^2\tau^2))u^{k+1}_{N} = \phi_1(2a^2\tau^2h + bh^2\tau^2) + \alpha_1(2u^{k}_N h^2 - u^{k-1}_N h^2 + f(l, t)\tau^2h^2 + 0.5d(u^{k-1}_N - 4u^{k}_N)\tau h^2)$

```python
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact, interactive, fixed, interact_manual
from IPython.display import display
import ipywidgets as widgets
```

```python
params = {
    'a': 1.0,
    'b': 1.0,
    'c': -1.0,
    'd': 3.0,
    'f': lambda x, t: -np.cos(x) * np.exp(-t),
    'alpha_0': 1.0,
    'beta_0': 0.0,
    'f_phi_0': lambda t: np.exp(-t),
    'alpha_1': 1.0,
    'beta_1': 0.0,
    'f_phi_1': lambda t: -np.exp(-t),
    'f_psi_0': lambda x: np.sin(x),
    'f_psi_1': lambda x: -np.sin(x),
    'psi_0_der_1': lambda x: np.cos(x),
    'psi_0_der_2': lambda x: -np.sin(x),
    'interval': (0.0, np.pi),
    'analytical_solution': lambda x, t: np.exp(-t) * np.sin(x)
}
```

```python
class Task:
    def __init__(self, **params):
        self.a = params['a']
        self.b = params['b']
        self.c = params['c']
        self.d = params['d']
        self.f = params['f']
        
        self.f_psi_0 = params['f_psi_0']
        self.f_psi_1 = params['f_psi_1']
        self.interval = params['interval']
        
        self.alpha_0 = params['alpha_0']
        self.beta_0 = params['beta_0']
        self.f_phi_0 = params['f_phi_0']
        
        self.alpha_1 = params['alpha_1']
        self.beta_1 = params['beta_1']
        self.f_phi_1 = params['f_phi_1']
        
        self.psi_0_der_1 = params['psi_0_der_1']
        self.psi_0_der_2 = params['psi_0_der_2']
        
        self.analytical_solution = params['analytical_solution']
            
class Solver:
    
    @classmethod
    def solve(cls, task:Task, n, max_time, scheme_type, bound_appr_type, initial_appr_type):
        h, tau, time_cnt = cls.__prepare(task.interval[0], task.interval[1], task.a, n, max_time)
        if scheme_type == 'explicit':
            u = cls.__explicit_solve(task, h, tau, n, time_cnt, bound_appr_type, initial_appr_type)
        elif scheme_type == 'implicit':
            u = cls.__implicit_solve(task, h, tau, n, time_cnt, bound_appr_type, initial_appr_type)
        x = cls.__get_split_by_space(h, n)
        t = cls.__get_split_by_time(tau, time_cnt)
        return u, x, t
            
    @classmethod
    def __explicit_solve(cls, task:Task, h, tau, n, time_cnt, bound_appr_type, initial_appr_type):
        u = np.zeros((time_cnt, n))
        z_0 = z_1 = b_0 = c_0 = a_n = b_n = 0.0
        if bound_appr_type == '2p2a':
            z_0 = 2 * task.a ** 2 * tau ** 2 * h - task.b * h ** 2 * tau ** 2
            z_1 = 2 * task.a ** 2 * tau ** 2 * h + task.b * h ** 2 * tau ** 2
            b_0 = task.alpha_0 * (-2 * task.a ** 2 * tau ** 2 - \
                  h ** 2 + task.c * tau ** 2 * h ** 2 - task.d * tau * h ** 2 / 2.0) + task.beta_0 * z_0
            c_0 = task.alpha_0 * 2 * task.a ** 2 * tau ** 2
            a_n = -task.alpha_1 * 2 * task.a ** 2 * tau ** 2
            b_n = task.alpha_1 * (2 * task.a ** 2 * tau ** 2 + \
                  h ** 2 - task.c * tau ** 2 * h ** 2 + task.d * tau * h ** 2 / 2.0) + task.beta_1 * z_1
            
        for j in range(n):
            u[0][j] = task.f_psi_0(h * j)
        
        if initial_appr_type == '2p1a':
            for j in range(n):
                u[1][j] = task.f_psi_0(h * j) + tau * task.f_psi_1(h * j)
        elif initial_appr_type == '2p2a':
            for j in range(n):
                u[1][j] = (1 + task.c * tau ** 2 / 2) * task.f_psi_0(h * j) + \
                          (task.b * tau ** 2 / 2) * task.psi_0_der_1(h * j) + \
                          (tau - task.d * tau ** 2 / 2) * task.f_psi_1(h * j) + \
                          (task.a ** 2 * tau ** 2 / 2) * task.psi_0_der_2(h * j) + \
                          (tau ** 2 / 2) * task.f(h * j, 0.0)
                
        for k in range(2, time_cnt):
            for j in range(1, n - 1):
                u[k][j] = ((2 * cls.__sigma ** 2 - task.b * tau ** 2 / h) * u[k - 1][j - 1] + \
                          (2 * tau ** 2 * task.c - 4 * cls.__sigma ** 2 + 4) * u[k - 1][j] + \
                          (2 * cls.__sigma ** 2 + task.b * tau ** 2 / h) * u[k - 1][j + 1] + \
                          (2 * tau ** 2 * task.f(h * j, tau * (k - 1))) - \
                          (2 - task.d * tau) * u[k - 2][j]) / (2 + task.d * tau)
            if bound_appr_type == '2p1a':
                u[k][0] = (h * task.f_phi_0(tau * k) - task.alpha_0 * u[k][1]) / (h * task.beta_0 - task.alpha_0)
                u[k][-1] = (h * task.f_phi_1(tau * k) + task.alpha_1 * u[k][-2]) / (h * task.beta_1 + task.alpha_1)
            elif bound_appr_type == '3p2a':
                u[k][0] = (2 * h * task.f_phi_0(tau * k) + task.alpha_0 * u[k][2] - 4 * task.alpha_0 * u[k][1]) / \
                          (-3 * task.alpha_0 + 2 * h * task.beta_0)
                u[k][-1] = (2 * h * task.f_phi_1(tau * k) - task.alpha_1 * u[k][-3] + 4 * task.alpha_1 * u[k][-2]) / \
                           (3 * task.alpha_1 + 2 * h * task.beta_1)
            elif bound_appr_type == '2p2a':
                d_0 = task.f_phi_0(tau * k) * z_0 + task.alpha_0 * (-2 * u[k - 1][0] * h ** 2 + u[k - 2][0] * h ** 2 - \
                      task.f(0.0, tau * k) * tau ** 2 * h ** 2 - task.d * u[k - 2][0] * tau * h ** 2 / 2.0)
                u[k][0] = (d_0 - c_0 * u[k][1]) / b_0
                d_n = task.f_phi_1(tau * k) * z_1 + task.alpha_1 * (2 * u[k - 1][-1] * h ** 2 - u[k - 2][-1] * h ** 2 + \
                      task.f(task.interval[1], tau * k) * tau ** 2 * h ** 2 + task.d * u[k - 2][-1] * tau * h ** 2 / 2.0)
                u[k][-1] = (d_n - a_n * u[k][-2]) / b_n
        return u
    
    @classmethod
    def __implicit_solve(cls, task:Task, h, tau, n, time_cnt, bound_appr_type, initial_appr_type):
        u = np.zeros((time_cnt, n))
        d = np.zeros((time_cnt, n))
        
        a = np.array([0.0] + [task.b * tau ** 2 * h - 2.0 * task.a ** 2 * tau ** 2 for _ in range(1, n - 1)] + [0.0])
        b = np.array([0.0] + [2.0 * h ** 2 + 4.0 * task.a ** 2 * tau ** 2 - 2.0 * task.c * tau ** 2 * h ** 2 + \
                              3.0 * task.d * tau * h ** 2 for _ in range(1, n - 1)] + [0.0])
        c = np.array([0.0] + [-task.b * tau ** 2 * h - 2.0 * task.a ** 2 * tau ** 2 for _ in range(1, n - 1)] + [0.0])
        
        if bound_appr_type == '2p1a':
            b[0] = h * task.beta_0 - task.alpha_0
            c[0] = task.alpha_0
            a[-1] = -task.alpha_1
            b[-1] = task.alpha_1 + h * task.beta_1
        elif bound_appr_type == '3p2a':
            b[0] = -3 * task.alpha_0 + 2 * h * task.beta_0 + a[1] * task.alpha_0 / (-task.b * tau ** 2 * h - 2 * task.a ** 2 * tau ** 2)
            c[0] = 4 * task.alpha_0 + b[1] * task.alpha_0 / (-task.b * tau ** 2 * h - 2 * task.a ** 2 * tau ** 2)
            a[-1] = -4 * task.alpha_1 - b[1] * task.alpha_1 / (task.b * tau ** 2 * h - 2 * task.a ** 2 * tau ** 2)
            b[-1] = 3 * task.alpha_1 + 2 * h * task.beta_1 - c[1] * task.alpha_1 / (task.b * tau ** 2 * h - 2 * task.a ** 2 * tau ** 2)
        elif bound_appr_type == '2p2a':
            z_0 = 2.0 * task.a ** 2 * tau ** 2 * h - task.b * h ** 2 * tau ** 2
            z_1 = 2.0 * task.a ** 2 * tau ** 2 * h + task.b * h ** 2 * tau ** 2
            b[0] = task.alpha_0 * (-2.0 * task.a ** 2 * tau ** 2 - \
                   h ** 2 + task.c * tau ** 2 * h ** 2 - 3 * task.d * tau * h ** 2 / 2.0) + task.beta_0 * z_0
            c[0] = task.alpha_0 * 2.0 * task.a ** 2 * tau ** 2
            a[-1] = -task.alpha_1 * 2.0 * task.a ** 2 * tau ** 2
            b[-1] = task.alpha_1 * (2.0 * task.a ** 2 * tau ** 2 + \
                    h ** 2 - task.c * tau ** 2 * h ** 2 + 3 * task.d * tau * h ** 2 / 2.0) + task.beta_1 * z_1
            
        for j in range(n):
            u[0][j] = task.f_psi_0(h * j)
        
        if initial_appr_type == '2p1a':
            for j in range(n):
                u[1][j] = task.f_psi_0(h * j) + tau * task.f_psi_1(h * j)
        elif initial_appr_type == '2p2a':
            for j in range(n):
                u[1][j] = (1.0 + task.c * tau ** 2 / 2) * task.f_psi_0(h * j) + \
                          (task.b * tau ** 2 / 2) * task.psi_0_der_1(h * j) + \
                          (tau - task.d * tau ** 2 / 2) * task.f_psi_1(h * j) + \
                          (task.a ** 2 * tau ** 2 / 2) * task.psi_0_der_2(h * j) + \
                          (tau ** 2 / 2) * task.f(h * j, 0.0)
                
        for k in range(2, time_cnt):
            for j in range(1, n - 1):
                d[k][j] = 2.0 * tau ** 2 * h ** 2 * task.f(h * j, tau * k) - \
                          task.d * tau * h ** 2 * u[k - 2][j] + \
                          4 * task.d * tau * h ** 2 * u[k - 1][j] - \
                          2 * h ** 2 * (u[k - 2][j] - 2 * u[k - 1][j])
            if bound_appr_type == '2p1a':
                d[k][0] = h * task.f_phi_0(tau * k)
                d[k][-1] = h * task.f_phi_1(tau * k)
            elif bound_appr_type == '3p2a':
                d[k][0] = 2.0 * h * task.f_phi_0(tau * k) + d[k][1] * task.alpha_0 / (-task.b * tau ** 2 * h - 2 * task.a ** 2 * tau ** 2)
                d[k][-1] = 2.0 * h * task.f_phi_1(tau * k) - d[k][-2] * task.alpha_1 / (task.b * tau ** 2 * h - 2 * task.a ** 2 * tau ** 2)
            elif bound_appr_type == '2p2a':
                d[k][0] = task.f_phi_0(tau * k) * z_0 + task.alpha_0 * (-2.0 * u[k - 1][0] * h ** 2 + u[k - 2][0] * h ** 2 - \
                          task.f(0.0, tau * k) * tau ** 2 * h ** 2 - task.d * (u[k - 2][0] - 4 * u[k - 1][0]) * tau * h ** 2 / 2.0)
                d[k][-1] = task.f_phi_1(tau * k) * z_1 + task.alpha_1 * (2.0 * u[k - 1][-1] * h ** 2 - u[k - 2][-1] * h ** 2 + \
                           task.f(task.interval[1], tau * k) * tau ** 2 * h ** 2 + task.d * (u[k - 2][-1] - 4 * u[k - 1][-1]) * tau * h ** 2 / 2.0)
            u_k = cls.__tma(a, b, c, d, n, k)
            for j in range(n):
                u[k][j] = u_k[j]
        return u
    
    @classmethod
    def __tma(cls, a, b, c, d, n, k):
        A = np.zeros(n)
        B = np.zeros(n)
        x = np.zeros(n)
        A[0] = -c[0] / b[0]
        B[0] = d[k][0] / b[0]
        
        for j in range(1, n):
            A[j] = -c[j] / (b[j] + a[j] * A[j - 1])
            B[j] = (d[k][j] - a[j] * B[j - 1]) / (b[j] + a[j] * A[j - 1])
        x[-1] = B[-1]
        for j in range(n - 2, -1, -1):
            x[j] = A[j] * x[j + 1] + B[j]
        return x
    
    @classmethod
    def analytical_solution(cls, task:Task, n, max_time):
        h, tau, time_cnt = cls.__prepare(task.interval[0], task.interval[1], task.a, n, max_time)
        res = np.zeros((time_cnt, n))
        for k in range(time_cnt):
            for j in range(n):
                res[k][j] = task.analytical_solution(h * j, tau * k)
        return res
    
    @classmethod
    def __get_split_by_space(cls, h, n):
        return np.array([h * j for j in range(n)])
    
    @classmethod
    def __get_split_by_time(cls, tau, time_cnt):
        return np.array([tau * k for k in range(time_cnt)])
    
    @classmethod
    def __prepare(cls, begin, end, a, n, max_time):
        h = (end - begin) / (n - 1)
        tau = cls.__sigma * h / a
        time_cnt = int(max_time / tau)
        return h, tau, time_cnt
        
    __sigma = 0.4
```
